```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('blue_image.jpg')

print ("The type of image is ", type(image), "with dimensions ",image.shape)
```

```python
%matplotlib inline

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB) 

plt.imshow(image_copy)
```

```python
lower_blue = np.array([0,0,170])
upper_blue = np.array([60,75,230])
```

```python
mask = cv2.inRange(image_copy, lower_blue, upper_blue)

plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image_copy)

masked_image[mask!=0] = [0, 0, 0]

plt.imshow(masked_image)
```

```python
background_image = cv2.imread('space.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB) 

crop_background = background_image[0:540, 0:960]
crop_background[mask == 0] = [0, 0, 0]

plt.imshow(crop_background)
```

```python
complete_image = crop_background + masked_image

plt.imshow(complete_image)
```

```python
complete_image = cv2.cvtColor(complete_image, cv2.COLOR_BGR2RGB) 
cv2.imwrite('my_graphic.jpg',complete_image)
```
